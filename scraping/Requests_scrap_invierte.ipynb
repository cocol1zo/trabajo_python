```python
import requests
from bs4 import BeautifulSoup
from csv import writer
# ver video
# https://www.youtube.com/watch?v=4UcqECQe5Kc
```

```python
url = 'http://ofi5.mef.gob.pe/invierte/ejecucion/verFichaEjecucion/2428470'
page = requests.get(url)
soup = BeautifulSoup(page.text , 'html.parser')
print(soup)
# soup = BeautifulSoup(page.content, 'html.parser')
# print(soup)
```

# Directo

```python
print(soup.prettify())
```

# Find

```python
el = soup.find('div')
print(el)
```

# find_all() or find_All()

```python
el = soup.find_all('div')
print(el)
```

```python
el = soup.find_all('div')[1]
print(el)
```

```python
# soup.find(id='section-1')
el = soup.find(id='divVistaPreliminar')
print(el)
```

```python
el = soup.find(class_='divVistaPreliminar')
print(el)
```

# Select

```python
el = soup.select('#divVistaPreliminar')
print(el)
```

```python
el = soup.select('#divVistaPreliminar')[0]
print(el)
```

# Get_text

```python
el = soup.find(class_='text-primary text-center').get_text()
print(el)
```
