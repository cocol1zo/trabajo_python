```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://peru.as.com/resultados/futbol/torneo_descentralizado/2019/clasificacion/'
page = requests.get(url)
#src = resultado.content
#soup = BeautifulSoup(src, 'lxml')
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup)
```

# encuentra equipos

```python
eq = soup.findAll('span', class_='nombre-equipo')
#eq = soup.findAll("span",{"class":"nombre-equipo"})
len(eq)
```

```python
print(eq)
```

# lista equipos

```python
equipos = list()
count = 0
for i in eq:
    if count <18:
        equipos.append(i.text)
    else:
        break
    count += 1
    
print(equipos, len(equipos))
```

# puntos

```python
pt = soup.findAll('td', class_='destacado')
#eq = soup.findAll("span",{"class":"nombre-equipo"})
len(eq)

puntos = list()

count = 0
for i in pt:
    if count <18:
        puntos.append(i.text)
    else:
        break
    count += 1
#df = pd.DataFrame({'Nombre' : equipos, 'Puntos' :puntos}, index=list(range(1,19)))
#print(df)
```

# crear dataframe con dos columnas

```python
#df = pd.DataFrame({'Nombre' : equipos, 'Puntos' :puntos}, index=list(range(1,19)))

data = pd.DataFrame({"Nombre" : equipos, "Puntos" : puntos}, index=list(range(1,19)))
print(data)
```

# crear a CSV

```python
data.to_csv('Clasificacion_futbol_peruano.csv', index=False)
```
